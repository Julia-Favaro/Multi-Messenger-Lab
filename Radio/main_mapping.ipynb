### IMPORTIAMO LE LIBRERIE

```python
import os 
import sys
import numpy as np

import matplotlib.pyplot as plt
from scipy import optimize

sys.path.append("/home/jovyan/experience-milky-way-at-21cm-2023-mmlab02")
from Radiofitter_j import Radio_fitter
```

### SECONDO QUADRANTE

```python
#raw data
II_quad=[]
for ii in range(70,86): #87
    name="spectrum_662"+str(ii)+".txt"
    II_quad=np.append(II_quad, name)
    
for ii in range(len(II_quad)):
    input_filename=II_quad[ii]
    mg_fit=Radio_fitter("relative velocity to LSR[km/s]", "intensity [arb. un.]")
    mg_fit.read_data(input_filename)
    #mg_fit.plot_data()
```

```python
#preprocessing
beg=[-120,-120,-120,-120,-120,-120,-120,-120,-120,-100,-100,-100,-80,-80,-80,-40,-40,-40]#you get this looking at the data 
end=np.ones(len(II_quad))*20

for ii in range(len(II_quad)):
    input_filename=II_quad[ii]
    mg_fit=Radio_fitter("relative velocity to LSR[km/s]", "intensity [arb. un.]")
    mg_fit.read_data(input_filename)
    
    #after looking at the raw data
    mg_fit.baseline(beg[ii],end[ii])
    #mg_fit.preprocessing(plot=True)
```

```python
#fitting

x=[] #nel primo quadrante la distanza dal GC  univocamente determinata 
y=[]
xdeg=[]
ydeg=[]

dx=[]
dy=[]
dxdeg=[]
dydeg=[]

for ii in range(len(II_quad)):
    input_filename=II_quad[ii]
    mg_fit=Radio_fitter("relative velocity to LSR[km/s]", "intensity [arb. un.]")
    mg_fit.read_data(input_filename)
    
    mg_fit.baseline(beg[ii],end[ii])
    mg_fit.fit_data()
    mg_fit.plot_res()
    #mg_fit.print_fit()
    
    xfit,yfit,xdegfit,ydegfit,dxfit,dyfit,dxdegfit,dydegfit=mg_fit.distance_GC()
    x=np.append(x,xfit)
    y=np.append(y,yfit)
    xdeg=np.append(xdeg,xdegfit)
    ydeg=np.append(ydeg,ydegfit)
    
    dx=np.append(dx,dxfit)
    dy=np.append(dy,dyfit)
    dxdeg=np.append(dxdeg,dxdegfit)
    dydeg=np.append(dydeg,dydegfit)
```

### PRIMO QUADRANTE

```python
#raw data
I_quad=["spectrum_66199.txt","spectrum_66192.txt","spectrum_66193.txt","spectrum_66194.txt","spectrum_66195.txt", "spectrum_66288.txt","spectrum_66289.txt","spectrum_66290.txt","spectrum_66291.txt"]
for ii in range(74,95):
    name="spectrum_663"+str(ii)+".txt"
    I_quad=np.append(I_quad, name)
    
for ii in range(len(I_quad)):
    input_filename=I_quad[ii]
    mg_fit=Radio_fitter("relative velocity to LSR[km/s]", "intensity [arb. un.]")
    mg_fit.read_data(input_filename)
    #mg_fit.plot_data()
```

```python
#preprocessing

beg=[-60,-80,-80,-100,-90,-100,-100,-100,-100] #you get this looking at the data
end=[100, 90,100,60,60,25,25,25,25]
beg=np.append(beg,[-80,-60,-60,-60,-60,-60,-80,-80,-80,-80,-100,-100,-100,-100,-100,-100,-100,-100,-100,-100,-100])
end=np.append(end,[120,120,120,120,120,120,120,100,100,80,80,80,60,60,40,40,40,40,40,40,40])
for ii in range(len(I_quad)):
    input_filename=I_quad[ii]
    mg_fit=Radio_fitter("relative velocity to LSR[km/s]", "intensity [arb. un.]")
    mg_fit.read_data(input_filename)
    
    #after looking at the raw data
    mg_fit.baseline(beg[ii],end[ii])
    #mg_fit.preprocessing(plot=True)
```

```python
#fitting

for ii in range(len(I_quad)):
    input_filename=I_quad[ii]
    mg_fit=Radio_fitter("relative velocity to LSR[km/s]", "intensity [arb. un.]")
    mg_fit.read_data(input_filename)
    
    mg_fit.baseline(beg[ii],end[ii])
    mg_fit.fit_data()
    mg_fit.plot_res()
    
    xfit,yfit,xdegfit,ydegfit,dxfit,dyfit,dxdegfit,dydegfit=mg_fit.distance_GC()
    x=np.append(x,xfit)
    y=np.append(y,yfit)
    xdeg=np.append(xdeg,xdegfit)
    ydeg=np.append(ydeg,ydegfit)
    
    dx=np.append(dx,dxfit)
    dy=np.append(dy,dyfit)
    dxdeg=np.append(dxdeg,dxdegfit)
    dydeg=np.append(dydeg,dydegfit)
```

### MAPPA DELLA VIA LATTEA

```python
plt.figure(figsize=(12,12))
plt.title("Milky Way galaxy in 21 cm wavelength",fontsize = 30)
plt.xlabel("x [kpc]",fontsize = 30)
plt.ylabel("y [kpc]",fontsize = 30)
plt.minorticks_on()

plt.errorbar(x, y, dy, dx, fmt='.',color='blue',label='No degeneration')
plt.errorbar(xdeg,ydeg,dydeg, dxdeg, fmt='.', color='red',label='degeneration')
plt.errorbar(0,8.5,fmt='*',markersize=10, color='orange', label='Sun')
plt.axhline(0, linestyle='dashed', color='black', alpha=0.25)
plt.axvline(0,linestyle='dashed', color='black', alpha=0.25)
plt.legend(loc='best',fontsize = 25)
plt.ylim([-20,20])
plt.xlim([-20,20])

main_dir = os.getcwd() #define work, results directories and path of the output file
results_julia_dir = os.path.join(main_dir,"results_julia")
output_filename = os.path.join(results_julia_dir,"Galaxy mapping")
plt.savefig(output_filename+".png")
```
