La tecnica più utilizzata per rimuovere le fonti dalle statistiche delle immagini è chiamata **sigma clipping**. In breve, i pixel che sono al di sopra o al di sotto di un multiplo di sigma dalla mediana vengono scartati e le statistiche vengono ricalcolate. La procedura viene tipicamente ripetuta per un certo numero di iterazioni o fino a convergenza.

La conversione dell'FWHM da pixel ad arcosecondi (utilizzando la scala dei pixel) determina il **seeing** in quelle condizioni astronomiche.

## **Possible sources of errors for the counts in aperture fotometry:**

Aperture photometry is a technique used in astronomy to measure the flux of light from an astronomical object within a specified aperture. When performing aperture photometry, several sources of error can affect the accuracy of the measurements. Here are the main types of errors to consider:

### 1. Photon (Poisson) Noise
This error arises from the discrete nature of photons and follows a Poisson distribution. It is inherent to the counting statistics of the photons collected from the source.

$\sigma_{\text{photon}} = \sqrt{N_{\text{source}}}$

### 2. Sky Background Noise
This error is due to the variation in the sky background within the aperture. The sky background includes contributions from scattered light, moonlight, airglow, and other sources.

$\sigma_{\text{sky}} = \sqrt{N_{\text{sky}} \cdot A}$

where $(N_{\text{sky}}$ is the average number of photons per pixel from the sky background and $A$ is the area of the aperture in pixels.

### 3. Readout Noise
This is the electronic noise introduced by the detector during the readout process. It is usually specified by the detector's manufacturer and it is proportional to the sqrt of the area of the aperture in pixels.

### 4. Flat-Fielding Errors (considered in the image calibration)
Flat-fielding corrects for pixel-to-pixel sensitivity variations. Errors in flat-fielding can introduce additional noise if the flat-field calibration is imperfect.

### 5. Dark Current Noise
This is the noise from the dark current, which is the charge accumulated in the absence of light due to thermal electrons. This is usually subtracted during calibration, but residuals can still contribute to the noise.

### 6. Aperture Placement Error (doing our best)
Errors can occur from inaccurate placement of the aperture, leading to loss or inclusion of unwanted light. This can be minimized by careful centroiding of the target object.

### 7. Atmospheric Effects (not considered)
Atmospheric turbulence (seeing) and extinction can introduce variability in the measured flux. This is usually addressed by differential photometry, comparing the target star with nearby reference stars.
